```python
from collections import OrderedDict
def exact_top_k(doc_dict, rank_score, k):
    relevance_scores = {}
    i = 0
    for doc_id in doc_dict.keys():
        relevance_scores[doc_id] = rank_score[i]
        i = i + 1
        
    sorted_value = OrderedDict(sorted(relevance_scores.items(), key = lambda x: x[1], reverse = True))
    top_k = {j: sorted_value[j] for j in list(sorted_value)[:k]}
    return top_k

#misal k = 3
top_3=exact_top_k(doc_dict,doc_scores,3)

print(top_3)
```

```python
import numpy as np

#menghitung df yang merupakan freq kata pada suatu dokumen 
def wordDocFre(vocab,news):
    df={}
    for word in vocab:
        frq = 0
        for x in news.values():
            if word in tokenisasi(x):
                frq=frq+1
        df[word]=frq 
    return df

#print(wordDocFre(vocab,news))
```

```python
def inverseDocFre(vocab,doc_fre,length):
    idf={}
    for word in vocab:
        idf[word]=idf[word]=1+np.log10((length+1)/(doc_fre[word]+1))
    return idf

#print(inverseDocFre(vocab,wordDocFre(vocab,doc_dict),len(doc_dict)))
idf = inverseDocFre(vocab,wordDocFre(vocab,news),len(news))
print(idf)
```

```python
def tfidf(vocab,tf,idf_scr,news):
    tf_idf_scr={}
    for news_id in news.keys():
        tf_idf_scr[news_id]={}
    for word in vocab:
        for news_id, x in news.items():
            tf_idf_scr[news_id][word] = tf[news_id][word]*idf_scr[word]
    return tf_idf_scr
#tem document
tf_idf = tfidf(vocab, termFrequencyInDoc(vocab,news),inverseDocFre(vocab,wordDocFre(vocab,news),len(news)),news)
print(tf_idf)
TD = np.zeros((len(vocab),len(news)))
for word in vocab:
    for news_id, x in tf_idf.items():
        ind1=vocab.index(word)
        ind2=list(tf_idf.keys()).index(news_id)
        TD[ind1][ind2]=tf_idf[news_id][word]

print("TD:",TD[:,0])
```

```python
import math
def cosine_sim(vec1,vec2):
    vec1=list(vec1)
    vec2=list(vec2)
    dot_prod=0
    for i, v in enumerate(vec1):
        dot_prod+=v*vec2[i]
    mag_1=math.sqrt(sum([x**2 for x in vec1]))
    mag_2=math.sqrt(sum([x**2 for x in vec2]))
    return dot_prod/(mag_1 * mag_2)

print(cosine_sim(TQ[:, 0], TD[:, 0])) #query & doc1
print(cosine_sim(TQ[:, 0], TD[:, 1])) #query & doc2
print(cosine_sim(TQ[:, 0], TD[:, 2])) #query & doc3
print(cosine_sim(TQ[:, 0], TD[:, 3])) #query & doc4
print(cosine_sim(TQ[:, 0], TD[:, 4])) #query & doc5
```

```python
from collections import OrderedDict
def exact_top_k(news, TD, q, k):
    relevance_scores = {}
    i = 0
    for news_id in news.keys():
        relevance_scores[news_id] = cosine_sim(q, TD[:, i])
        i = i + 1
    sorted_value = OrderedDict(sorted(relevance_scores.items(), key=lambda x: x[1], reverse = True))
    top_k = {j: sorted_value[j] for j in list(sorted_value)[:k]}
    return top_k
top_2 = exact_top_k(news, TD, TQ[:, 0], 2)
#misal k = 3
top_5=exact_top_k(news,TD,TQ[:,0],5)
print(top_5)
```
